```python
#Making first population by reading from test file
# Size of chromosome = 8(9 to 5) * number of rooms * 5(days (MON-FRI))
#[[DAY  [ROOM 1  [STARTING HOUR   [][][][]...  ]...   ]...   ]...]

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
rooms = list(range(1, roomCount+1))
startTime = np.arange(8.5, 17, 0.5)

day_room_time = []
for d in days:
    for r in rooms:
        for s in startTime:
            
            val=""
            val+=(str(d))
            val+=("_")
            val+=(str(r))
            val+=("_")
            val+=(str(s))
            day_room_time.append(val)
```

```python
#All classes for all batches(groups)+section+Teacher+duration
classes = []

for i in range(0, totalRows, 2):
    batch = df.values[rowList[i],0]
    for j in range(1, totalColumns):
        
        if df.values[rowList[i+1],j] == "1":
#             print("1 hour class")
            classes.append(batch+"_"+df.values[rowList[i],j]+"_"+df.values[rowList[i+1],j])
            classes.append(batch+"_"+df.values[rowList[i],j]+"_"+df.values[rowList[i+1],j])
            classes.append(batch+"_"+df.values[rowList[i],j]+"_"+df.values[rowList[i+1],j])

        else:
#             print("1.5 hour class")
            classes.append(batch+"_"+df.values[rowList[i],j]+"_"+df.values[rowList[i+1],j])
            classes.append(batch+"_"+df.values[rowList[i],j]+"_"+df.values[rowList[i+1],j])
print("len(classes)", len(classes))

if totalClassesInWeek < len(classes):
    print("CONDITIONS NOT ACCEPTABLE")
```

```python
#Now randomly placing classes in chromosome and adding to population array
population = []
populationSize = 500

for i in range(populationSize):
    chromosome = [""] * len(day_room_time)

    for _class in classes:

        #generate a random empty position in chromosome to place class into
        pos = randint(0, len(day_room_time)-1)
        while chromosome[pos] != "":
            pos = randint(0, len(day_room_time)-1)

        chromosome[pos] = _class
    population.append(chromosome) 
```

```python
def calculateFitness(chromosome):
    
    softConstraintDelta = 1
    hardConstraintDelta = 5
    
    
    day_room = []
    fitness = 5000
    
    for i in range(len(chromosome)):
        val = chromosome[i]
        idx = day_room_time[i]
        
        if val != "":
            
#             print(val)
            splits = val.split("_")
#             print(splits)
            group = splits[0] + "_" + splits[1]
            teacher = splits[2]
            duration = splits[3]

            splits1 = idx.split("_")
            day = splits1[0]
            room = splits1[1]
            startTime = splits1[2]

            #Duration(Room occupied) checks
            if duration == "1":
                if startTime != "16.5":
                    if i < 1272:
                        if chromosome[i+1] != "":
                            fitness -= hardConstraintDelta
                        if chromosome[i+2] != "":
                            fitness -= hardConstraintDelta

            elif duration == "1.5":
                if startTime != "16.5":
                    if i < 1271:
                        if chromosome[i+1] != "":
                            fitness -= hardConstraintDelta
                        if chromosome[i+2] != "":
                            fitness -= hardConstraintDelta
                        if chromosome[i+3] != "":
                            fitness -= hardConstraintDelta
                            
            #Friday prayer checks(no classes b/w 1-2)
            if day == "Fri":
                if duration == "1":
                    if startTime == "12.5":
                        fitness -= hardConstraintDelta
                    if startTime == "13":
                        fitness -= hardConstraintDelta
                    if startTime == "13.5":
                        fitness -= hardConstraintDelta
                        
                elif duration == "1.5":
                    if startTime == "12":
                        fitness -= hardConstraintDelta
                    if startTime == "12.5":
                        fitness -= hardConstraintDelta
                    if startTime == "13":
                        fitness -= hardConstraintDelta
                    if startTime == "13.5":
                        fitness -= hardConstraintDelta
                        
#             print(teacher)
#             print(startTime)
#             print(duration)
            #Teacher constraint(one teacher can not teach more than 1 class at a time)
            #room repeat after 17(0-16 is one room then 2nd and so on...)
            for r in range(roomCount):
                
                if duration == "1":
                    for j in range(3):
                        
                        val_ = chromosome[(r*17)+j]
                    
                        if val_ != "":
                            print
                            teacherCompare = val_.split("_")[2]
                            
                            if j != 2:
                                if teacher == teacherCompare:
                                    fitness -= hardConstraintDelta
                            else:#soft constraint check (consecutive classes of a teacher)
                                if teacher == teacherCompare:
                                    fitness -= softConstraintDelta
                elif duration == "1.5":
                    for j in range(4):

                        val_ = chromosome[(r*17)+j]

                        if val_ != "":
                            teacherCompare = val_.split("_")[2]
                            if j != 3:
                                if teacher == teacherCompare:
                                    fitness -= hardConstraintDelta
                            else:#soft constraint check (consecutive classes of a teacher)
                                if teacher == teacherCompare:
                                    fitness -= softConstraintDelta
            
            
            
            #Teacher preferences
            for tp in teacherPreference:
                t = tp[0]
                p = tp[1]
                if teacher == t:
                    if p == "m":
                        if float(startTime) > 13:
                            fitness -= softConstraintDelta
                    elif p == "e":
                        if float(startTime) < 13:
                            fitness -= softConstraintDelta
                    
            
    return fitness/5000   
```

```python
def mutation(chromosome, levelOfRandomness = 5):
    
    for i in range(levelOfRandomness):
        
        chromosomeIdx = randint(0, len(chromosome)-1)
        classIdx = randint(0, len(classes))
        val = ""
        if classIdx != len(classes):
            val = classes[classIdx]
        chromosome[chromosomeIdx] = val
    return chromosome
```
